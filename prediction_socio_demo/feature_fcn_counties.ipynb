```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_combined_ratio = pd.read_csv('data_combined_ratio.csv')
data_combined_ratio.head()
```

```python
data_combined_ratio['population_ratio'] = (data_combined_ratio['DrugReports']/data_combined_ratio['popultion'])*100
data_combined_ratio.head()
```

```python
Heroin_data = data_combined_ratio.loc[(data_combined_ratio['SubstanceName'] == 'Oxycodone')]
Heroin_data.head()
```

```python
Heroin_data_train = Heroin_data.loc[(Heroin_data['YYYY'] <= 2015)]
Heroin_data_test = Heroin_data.loc[(Heroin_data['YYYY'] == 2016)]

Heroin_data_test
```

```python
Heroin_data_train_x = Heroin_data_train.iloc[:, 10:102]
Heroin_data_train_x
```

```python
Heroin_data_train_y = Heroin_data_train.iloc[:, 102]
Heroin_data_train_y
```

```python
Heroin_data_test_x = Heroin_data_test.iloc[:, 10:102]
Heroin_data_test_y = Heroin_data_test.iloc[:, 102]
```

```python
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.linear_model import HuberRegressor
```

```python
Heroin_feature_chosen_10 = ['HC03_VC46','HC03_VC200','HC03_VC11','HC03_VC192','HC03_VC91','HC03_VC47','HC03_VC40','HC03_VC198','HC03_VC67','HC03_VC86']
```

```python
Heroin_data_train_x_chosen = Heroin_data_train_x[Heroin_feature_chosen_10]
Heroin_data_train_x_chosen
```

```python
Heroin_data_test_x_chosen = Heroin_data_test_x[Heroin_feature_chosen_10]
Heroin_data_test_x_chosen
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
regr_bayes = linear_model.BayesianRidge()
regr_bayes.fit(Heroin_data_train_x_chosen, Heroin_data_train_y)
print(f"test score:", regr_bayes.score(Heroin_data_test_x_chosen, Heroin_data_test_y))
```

```python
bayes_predict = regr_bayes.predict(Heroin_data_test_x_chosen)
print(mean_absolute_error(bayes_predict, np.array(Heroin_data_test_y)))
print(mean_squared_error(bayes_predict, np.array(Heroin_data_test_y)))
```

```python
regr_svr = SVR(kernel="rbf", C=1, gamma=0.1, epsilon=0.1)
regr_svr.fit(Heroin_data_train_x_chosen, Heroin_data_train_y)
print(f"test score:", regr_svr.score(Heroin_data_test_x_chosen, Heroin_data_test_y))
```

```python
svr_predict = regr_svr.predict(Heroin_data_test_x_chosen)
print(mean_absolute_error(svr_predict, np.array(Heroin_data_test_y)))
print(mean_squared_error(svr_predict, np.array(Heroin_data_test_y)))
```

```python
# FCN started here
import tensorflow as tf
outcome_mae = {}
outcome_mse = {}
```

```python
#fcn analysis
for i in range(10,21):
    for j in range(20,31):
        for k in [10]:
            model = tf.keras.Sequential(
                [
                    tf.keras.layers.Dense(10, name="layer1"),
                    tf.keras.layers.Dense(i, name="layer2"),
                    tf.keras.layers.Dense(j, name="layer3"),
                    tf.keras.layers.Dense(k, name="layer4"),
                    tf.keras.layers.Dense(4, name="layer5"),
                    tf.keras.layers.Dense(1, activation='sigmoid', name="layer6"),
                ]
            )

            model.compile(loss='binary_crossentropy', optimizer='adam')
            model.fit(Heroin_data_train_x_chosen, Heroin_data_train_y, epochs=100, batch_size=10)
            predictions = model.predict(Heroin_data_test_x_chosen)
            prediction_number = (Heroin_data_test["popultion"]*predictions.flatten())/100
            print("i is: " + str(i) + " j is: " + str(j) + " k is: " + str(k))
            print(mean_absolute_error(np.array(prediction_number), np.array(Heroin_data_test["DrugReports"])))
            print(mean_squared_error(np.array(prediction_number), np.array(Heroin_data_test["DrugReports"])))
            outcome_mae[(i,j,k)] = mean_absolute_error(np.array(prediction_number), np.array(Heroin_data_test["DrugReports"]))
            outcome_mse[(i,j,k)] = mean_squared_error(np.array(prediction_number), np.array(Heroin_data_test["DrugReports"]))
        
```

```python
print(outcome_mse)
```

```python
print(outcome_mae)
```

```python
X = []
Y = []
Z = []
```

```python
for key in outcome_mae:
    if key[2] == 10:
        X.append(key[0])
        Y.append(key[1])
        Z.append(outcome_mae[key])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(projection='3d')

ax.plot_trisurf(np.array(X), np.array(Y), np.array(Z), color='xkcd:sky blue')

ax.set_xlabel('#neurons in first hidden layer')
ax.set_ylabel('#neurons in second hidden layer')
ax.set_zlabel('MAE')

plt.xticks(size=12)
plt.yticks(size=12)
plt.title("MAE under Different Number of Neurons", size = 15)

ax.grid(False)
plt.show()
fig.savefig('mae.jpg')
```
