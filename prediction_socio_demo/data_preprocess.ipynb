```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('MCM_NFLIS_Data.xlsx',sheet_name='Data')
demo_2010 = pd.read_csv('ACS_10_5YR_DP02/ACS_10_5YR_DP02_with_ann.csv')
demo_2011 = pd.read_csv('ACS_11_5YR_DP02/ACS_11_5YR_DP02_with_ann.csv')
demo_2012 = pd.read_csv('ACS_12_5YR_DP02/ACS_12_5YR_DP02_with_ann.csv')
demo_2013 = pd.read_csv('ACS_13_5YR_DP02/ACS_13_5YR_DP02_with_ann.csv')
demo_2014 = pd.read_csv('ACS_14_5YR_DP02/ACS_14_5YR_DP02_with_ann.csv')
demo_2015 = pd.read_csv('ACS_15_5YR_DP02/ACS_15_5YR_DP02_with_ann.csv')
demo_2016 = pd.read_csv('ACS_16_5YR_DP02/ACS_16_5YR_DP02_with_ann.csv')
```

```python
#creating empty dataset
for i in range(4,10):
    data["HC01_VC0"+str(i)] = np.nan
for i in range(10,110):
    data["HC01_VC"+str(i)] = np.nan
```

```python
#merging data from 2010 to 2016
for index, row in data.iterrows():
    if row['YYYY'] == 2010:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2010.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2010.loc[demo_2010['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2010.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2010.loc[demo_2010['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue       
    if row['YYYY'] == 2011:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2011.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2011.loc[demo_2011['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2011.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2011.loc[demo_2011['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue        
    if row['YYYY'] == 2012:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2012.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2012.loc[demo_2012['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2012.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2012.loc[demo_2012['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue        
    if row['YYYY'] == 2013:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2013.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2013.loc[demo_2013['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2013.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2013.loc[demo_2013['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue        
    if row['YYYY'] == 2014:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2014.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2014.loc[demo_2014['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2014.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2014.loc[demo_2014['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue        
    if row['YYYY'] == 2015:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2015.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2015.loc[demo_2015['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2015.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2015.loc[demo_2015['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue        
    if row['YYYY'] == 2016:
        for i in range(4,10):
            if ("HC01_VC0"+str(i)) in demo_2016.columns:
                data.at[index, "HC01_VC0"+str(i)] = demo_2016.loc[demo_2016['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC0"+str(i)].values[0]
        for i in range(10,220):
            try:
                if ("HC01_VC"+str(i)) in demo_2016.columns:
                    data.at[index, "HC01_VC"+str(i)] = demo_2016.loc[demo_2016['GEO.id2'] == str(row['FIPS_Combined'])]["HC01_VC"+str(i)].values[0]
            except:
                continue
```

```python
data.head()
```

```python
data['popultion'] = 0
population = pd.read_csv('maps-county2011.csv')

for idx in data.index:
#     print(idx)
#     print(data.at[idx, 'FIPS_Combined'])
#     print(population.loc[population['00000'] == data.at[idx, 'FIPS_Combined']]['308,745,538'].values[0])
    data.at[idx, 'popultion'] = int(''.join(population.loc[population['00000'] == data.at[idx, 'FIPS_Combined']]['308,745,538'].values[0].split(',')))

```

```python
data['population_ratio'] = data['DrugReports']/data['popultion']
```

```python
data = data.drop(data.loc[data['YYYY'] == 2017].index)
data_nona = data.dropna(axis=1) = data.dropna(axis=1)
```

```python
data_nona.head()
```

```python
data_nona.drop(columns='population_ratio')
```

```python
data_nona.to_csv('data_combined_number.csv', index=False)  
```
