```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_excel('MCM_NFLIS_Data.xlsx',sheet_name='Data')
data.head()
```

```python
name = set(data['SubstanceName'])
name = list(name)
#  Create a dictionary to save SubstanceName Of DrugReports All worth ,
DrugReports_dick = {
   
    }
for i in name:
    DrugReports_dick[i] = sum(data[data['SubstanceName'] == i]['DrugReports'])
print(DrugReports_dick)
#  Then sort and extract the first five values 
DrugReports_dick = sorted(DrugReports_dick.items(), key=lambda item:item[1])
DrugReports_dic = dict(DrugReports_dick)
```

```python
import seaborn as sns
keys = list(DrugReports_dic.keys())
# get values in the same order as keys, and parse percentage values
vals = [float(DrugReports_dic[k]) for k in keys]
sns.set(rc={'figure.figsize':(16,8)})
sns.barplot(x=keys, y=vals)
```

```python
DrugReports_five = np.array(DrugReports_dick)[-5:]
plt.figure(figsize=(10,10))

plt.pie(DrugReports_five[:,1],
        autopct='%1.1f%%',
        shadow=False,
        startangle=150)
plt.title("Substance")
plt.savefig("Substance.png")
plt.legend(labels=DrugReports_five[:,0])
plt.show()
plt.savefig('five_pie.jpg')
```

```python
for idx in data.index:
#     print(idx)
#     print(data.at[idx, 'FIPS_Combined'])
#     print(population.loc[population['00000'] == data.at[idx, 'FIPS_Combined']]['308,745,538'].values[0])
    data.at[idx, 'popultion'] = int(''.join(population.loc[population['00000'] == data.at[idx, 'FIPS_Combined']]['308,745,538'].values[0].split(',')))

```

```python
data['population_ratio'] = data['DrugReports']/data['popultion']
```

```python
data = data.drop(data.loc[data['YYYY'] == 2017].index)
data
```

```python
data_nona = data.dropna(axis=1)
data_nona
```

```python
data_Heroin = data_nona.drop(data_nona.loc[data_nona['SubstanceName'] != 'Heroin'].index)
data_Heroin.head()
```

```python
columns = data_Heroin.columns
print(columns[10:38])
```

```python
Heroin_corr = []
for heading in columns[10:38]:
    column_1 = data_Heroin[heading]
    column_2 = data_Heroin["population_ratio"]
    correlation = column_2.corr(column_1)
    Heroin_corr.append(correlation)

print(Heroin_corr)
```

```python
df = pd.DataFrame({'lab':columns[10:38], 'Heroin_corr':Heroin_corr})
ax = df.plot.bar(x='lab', y='Heroin_corr', rot=0)
```
