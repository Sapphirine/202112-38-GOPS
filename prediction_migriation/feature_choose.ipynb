```python
data = data.drop(data.loc[data['YYYY'] == 2017].index)
```

```python
data_nona = data.dropna(axis=1)
data_nona
```

```python
data_nona['popultion'] = 0
population = pd.read_csv('maps-county2011.csv')

for idx in data_nona.index:
#     print(idx)
#     print(data.at[idx, 'FIPS_Combined'])
#     print(population.loc[population['00000'] == data.at[idx, 'FIPS_Combined']]['308,745,538'].values[0])
    data_nona.at[idx, 'popultion'] = int(''.join(population.loc[population['00000'] == data_nona.at[idx, 'FIPS_Combined']]['308,745,538'].values[0].split(',')))

data_nona['population_ratio'] = data_nona['DrugReports']/data_nona['popultion']
data_nona
```

```python
columns = data_nona.columns[10:101]
print(columns)
```

```python
data_Heroin = data_nona.drop(data_nona.loc[data_nona['SubstanceName'] != 'Heroin'].index)
Heroin_corr = []
for heading in columns:
    column_1 = data_Heroin[heading]
    column_2 = data_Heroin["population_ratio"]
    correlation = column_2.corr(column_1)
    Heroin_corr.append(correlation)

print(Heroin_corr)
```

```python
df = pd.DataFrame({'lab':columns, 'Heroin_corr':Heroin_corr})
ax = df.plot.bar(x='lab', y='Heroin_corr', rot=0)
```

```python
Heroin_top_10_idx = np.argsort(np.abs(Heroin_corr))[-10:]
print(Heroin_top_10_idx)
for idx in Heroin_top_10_idx:
    print(columns[idx])
```
